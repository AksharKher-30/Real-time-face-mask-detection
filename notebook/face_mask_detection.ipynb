```python
import pandas as pd
import matplotlib.pyplot as plt

# Load training logs
df = pd.read_csv('/Users/akshar/Documents/DL/face mask detection/runs/s_runs/s_train/face_mask_detector_s/results.csv')

# Clean column names (good practice)
df.columns = df.columns.str.strip()

# Plot Precision, Recall, and mAP@0.5
plt.figure(figsize=(12, 6))
plt.plot(df['metrics/precision(B)'], label='Precision')
plt.plot(df['metrics/recall(B)'], label='Recall')
plt.plot(df['metrics/mAP50(B)'], label='mAP@0.5')
plt.plot(df['metrics/mAP50-95(B)'], label='mAP@0.5:0.95')
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.title('Model Performance Metrics')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def smooth_curve(values, weight=0.8):
    """Applies exponential moving average smoothing."""
    smoothed = []
    last = values[0]
    for point in values:
        smoothed_val = last * weight + (1 - weight) * point
        smoothed.append(smoothed_val)
        last = smoothed_val
    return smoothed
```

```python
# Apply smoothing
box_loss_smoothed = smooth_curve(df['train/box_loss'])
val_box_loss_smoothed = smooth_curve(df['val/box_loss'])
mAP50_smoothed = smooth_curve(df['metrics/mAP50(B)'])

# === Plot: Smoothed Box Losses ===
plt.figure(figsize=(10, 5))
plt.plot(box_loss_smoothed, label='Train Box Loss (Smooth)')
plt.plot(val_box_loss_smoothed, label='Val Box Loss (Smooth)', linestyle='--')
plt.title('Smoothed Box Losses Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(mAP50_smoothed, label='mAP@0.5 (Smoothed)')
plt.title('Smoothed mAP@0.5 Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('mAP')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
